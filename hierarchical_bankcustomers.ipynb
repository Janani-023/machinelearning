```python
#Business problem

#Cluster the bank customers into different groups based on their behaviour(income,credit usage,transactions)
```

```python
b=b.drop("cluster_",axis=1)
```

```python
numb=b.select_dtypes(include=["int64","float64"])
numb
```

```python
from sklearn.preprocessing import StandardScaler 
```

```python
scaler=StandardScaler()
snumb=scaler.fit_transform(numb)
snumb
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model=AgglomerativeClustering(n_clusters=3,linkage="ward")
clusters=model.fit(snumb)
b["cluster"]=clusters

```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
plt.figure(figsize=(12,6))
dendrogram(linkage(snumb,method="ward"))
plt.title("dendrogram plot")
plt.xlabel("customers")
plt.ylabel("distance")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(snumb[:,0],snumb[:,1],c=clusters,cmap="rainbow")
plt.show()
```
