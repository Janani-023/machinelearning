```python
import numpy as np
import matplotlib.pyplot as plt

x = x_test.iloc[:, 0].values
y_actual = y_test.values
y_predicted = y_pred

idx = np.argsort(x)

x_sorted = x[idx]
y_actual_sorted = y_actual[idx]
y_pred_sorted = y_predicted[idx]
plt.scatter(x_sorted, y_actual_sorted, label="Actual Data")
plt.plot(x_sorted, y_pred_sorted, label="Best Fit Line")

plt.xlabel("X (Feature)")
plt.ylabel("Y (Target)")
plt.legend()
plt.show()
```
